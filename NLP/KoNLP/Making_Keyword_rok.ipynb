## 형태소 수정1 및 삭제2 
### 삭제 
- 주격 조사, 선어말 어미, 형용사 파생 접미사, 선어말 어미+선어말 어미

### 수정
- 종결 어미, 연결 어미, 관형형 전성 어미 → ‘음’

- 동사 파생 접미사+연결 어미, 형용사 파생 접미사+관형형, 전성 어미, 형용사 파생 접미사+연결 어미, 선어말 어미+종결어미, 형용사 파생 접미사+연결 어미 → ‘함’

- 수정
형용사 파생 접미사과 연결 어미가 합쳐진 것이 없기에 연결어미 앞에 형용사 파생 접미사 ’하’ 가 있다면 ('함', 'XSA+EC')으로 추가해준다.

※ 한 번 키워드화된 단어들이기에 삭제할 형태소가 적었지만 처음부터 분석한다면 삭제할 형태소가 더 많아질 거 같다.

```python
del_lst = '''JKS
JKC
JKG
JKO
JKB
JKV
JKQ
JX
JC
EP
SF
SE
SSO
SSC
SC
EP+EP''' #삭제 형태소 리스트

fix_lst1 = '''XSV+EC
XSA+ETM
XSA+EC
EP+EF
XSA+EC''' #수정 형태소 리스트 -> '함'

fix_lst2 = '''EF
EC
ETM''' #수정 형태소 리스트 -> '음'

del_lst = del_lst.split('\n')
fix_lst1 = fix_lst1.split('\n')
fix_lst2 = fix_lst2.split('\n')
```

```python
for i in range(10):
  tmp_txt = sillim_review_mecab.loc[i,'review']
  mecab.pos(tmp_txt)
```

```python
cnt = len(sillim_review_mecab['review'].to_list()) #len(value_lst)
for i in range(10):
  tmp_txt = sillim_review_mecab.loc[i,'review']
  tmp_lst = mecab.pos(tmp_txt)
  sillim_review_mecab.at[i,'Mecab0'] = tmp_lst
  Mecab_lst1 = [] #임시 리스트
  Mecab_lst2 = [] #임시 리스트
  Mecab_lst3 = [] #임시 리스트
  for j in tmp_lst:
    if j[1] not in del_lst: # 삭제 리스트에 없다면 추가 
      Mecab_lst1.append(j)
  # 아래 반복문으로 한번에 하지 않는 이유는 XSA+EC을 수정하는 과정에서 
  for k in range(len(Mecab_lst1)):
    tmp_tu = Mecab_lst1[k]
    if tmp_tu[1] == 'NNG': #형태소가 명사라면
      Mecab_lst2.append((tmp_tu[0],'NNG'))
      Mecab_lst3.append(" "+tmp_tu[0]+" ") # join 했을 때 명사만 띄어쓰기 하기위해 
    elif tmp_tu[1] not in fix_lst1+fix_lst2: #수정 리스트
      Mecab_lst2.append(tmp_tu)
      Mecab_lst3.append(tmp_tu[0])
    elif tmp_tu[1] == 'EC' and Mecab_lst1[k-1][0] =='하': #연결 어미와 파생형 접미사 '하'인 경우
      Mecab_lst2.append(('함', 'XSA+EC'))
      Mecab_lst3.append('함')
    elif tmp_tu[1] in fix_lst1: # 수정1
      Mecab_lst2.append(('함', tmp_tu[1]))
      Mecab_lst3.append('함')
    elif tmp_tu[1] in fix_lst2: # 수정2
      Mecab_lst2.append(('음', tmp_tu[1]))
      Mecab_lst3.append('음')

  sillim_review_mecab.at[i,'Mecab1'] = Mecab_lst2 #형태소 분류 튜플 살려서 추가
  sillim_review_mecab.at[i,'Mecab1_1'] = "".join(Mecab_lst3) # 문자열로 추가
```
