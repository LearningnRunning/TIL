```python
import pandas as pd
from os.path import join
from os import listdir
STATIC_PATH = "./data"
file_lst = listdir(STATIC_PATH)
for i,f in enumerate(file_lst):
    print(i, f)
```

```python
file_name = file_lst[0]
file_path = join(STATIC_PATH,file_name)
daum_df = pd.read_csv(file_path)
daum_df['platform'] = 'daum'
print(daum_df.info())
```

```python
daum_df = daum_df.dropna(subset=['Review']).info()
```

```python
file_name = file_lst[2]
file_path = join(STATIC_PATH,file_name)
watcha_df = pd.read_csv(file_path)
watcha_df['platform'] = 'watcha'
print(watcha_df.info())
```

```python
file_name = file_lst[4]
file_path = join(STATIC_PATH,file_name)
cgv_df = pd.read_csv(file_path)
cgv_df['platform'] = 'cgv'
print(cgv_df.info())
```

```python
korean_review_df = pd.concat([daum_df, cgv_df])
```

```python
from hanspell import spell_checker
from tqdm import tqdm
tqdm.pandas()

# Assuming korean_review_df is your DataFrame

def spell_check(text):
    text = text.replace('&', 'and').replace('+', 'and').replace('*', ' ')
    try:
        result = spell_checker.check(text)  # Call your spell checker function
        checked_txt = result.as_dict()['checked']
        return checked_txt
    except Exception as e:
        # Print the index and text when a ParseError occurs
        print(f"ParseError at index {korean_review_df[korean_review_df['Review'] == text].index[0]}: {text}")
        return text  # You can choose to return the original text or another value

# Apply the spell_check function to each value in the 'review' column
korean_review_df['spell_checked_Review'] = korean_review_df['Review'].progress_apply(spell_check)
```

```python
# Apply the spell_check function to each value in the 'review' column
watcha_df['spell_checked_Review'] = watcha_df['Review'].progress_apply(spell_check)
```
