## 이모티콘 제거

```python
def deEmojify(text): #이모티콘 제거 함수
    regrex_pattern = re.compile(pattern = "["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U0001F1A0-\U0001F9FF"  # flags (iOS)
                           "]+", flags = re.UNICODE)
    return regrex_pattern.sub(r'',text)
```

```python
output = []
# text 정제
for text in reviewer_reviews:
    text = re.sub(r"(\w+)-\n(\w+)", r"\1\2", text)   # 안녕-\n하세요 -> 안녕하세요
    text = re.sub(r"(?<!\n\s)\n(?!\s\n)", " ", text.strip()) # "인\n공\n\n지능펙\n토리 -> 인공지능펙토리
    text = re.sub(r"\n\s*\n", "\n\n", text) # \n버\n\n거\n\n킹\n -> 버\n거\n킹
    output.append(text)
```

```python
doc_chunks = []

for line in output:
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=2000, # 최대 청크 길이
        separators=["\n\n", "\n", ".", "!", "?", ",", " ", ""], #  텍스트를 청크로 분할하는 데 사용되는 문자 목록
        chunk_overlap=0, # 인접한 청크 간에 중복되는 문자 수
    )
    chunks = text_splitter.split_text(line)
    for i, chunk in enumerate(chunks):
        doc = Document(
            page_content=chunk, metadata={"page": i, "source": i}
        )
        doc_chunks.append(doc)
```

```python
doc_chunks
```

```python
embeddings = OpenAIEmbeddings()
index = Chroma.from_documents(doc_chunks, embeddings)
```

```python
# 프롬프트 예시
"""I want you to act as my Keyword generator. As an expert in summarizing reviews to create keywords, the process involves analyzing the product name, product description, and all reviews related to the product. 
Focus on moisturizing power(保湿力), lasting power(持続力), texture(塗り感), absorptivity(密着感) and stimulation(刺激感)  in their reviews. analyzes a product's description and reviews and assigns a score from 1 to 10 for each of the following categories: moisturizing power (保湿力), lasting power (持続力), texture (塗り感), absorptivity (密着感) and stimulation (刺激感). The system should consider the overall sentiment expressed in the reviews and generate accurate ratings for each category.
If you don't know the answer, just say you don't know and don't try to make up an answer.


product_name: {product}
info: {info}
review: {review}

Below is an example.
ここが良いポイント
["オススメ", "梱包が綺麗", "使いやすい", "ブラシが柔らかい", "しっとりした粉質"]
ここではないこと
["使いこなせていない", "扱いにくい大きさ", "黄色味が強い", "しっとりした粉質"]

{ "保湿力" : 8, "持続力" : 7, "塗り感" : 6, "密着感" : 3, "刺激感" : 1}

you only answer in Japanese.

product_name: {product}
product_description: {product_description}
product_review: {review}
""",
```

```python
system_template="""I want you to act as my keyword classifier. Keywords are fixed. As an expert in summarizing reviews to create keywords. You will analyze it and create concise, descriptive keyword summaries for the positive and negative sentiments expressed in the reviews. If you don't know the answer, say you don't know and don't try to make up an answer. You don't have to answer anything that isn't mentioned.
 Don't write a description.

Below is an example.
Positive: fun, beneficial
Negative: boredom

You only answer in Korean.

{summaries}
"""
messages = [
    SystemMessagePromptTemplate.from_template(system_template),
    HumanMessagePromptTemplate.from_template("{question}")
]
prompt = ChatPromptTemplate.from_messages(messages)


chain_type_kwargs = {"prompt": prompt}
bk_chain = RetrievalQAWithSourcesChain.from_chain_type(
    ChatOpenAI(temperature=0),
    chain_type="stuff",
    retriever=index.as_retriever(),
    chain_type_kwargs=chain_type_kwargs,
    # reduce_k_below_max_tokens=True
)
```

```python
result = bk_chain({"question": output[0]})

print(f"질문 : {result['question']}")
print()
print(f"답변 : {result['answer']}")
```
