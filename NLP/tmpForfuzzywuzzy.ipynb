```python
import pandas as pd
from glob import glob
from fuzzywuzzy import fuzz
from tqdm import tqdm

data_path = './data/*.csv'
data_lst = glob(data_path)

# # 매핑할 사이트 데이터
# amazon_df = pd.read_excel('/Users/seongrok.kim/Github/TIL/NLP/data/amazon0809_ranking_areum.xlsx', index_col=0)

# # 체코 데이터
# checco_df = pd.read_excel('/Users/seongrok.kim/Github/TIL/NLP/data/checco0809_ranking_areum.xlsx', index_col=0)
# checco_df.reset_index(inplace=True)
# amazon_df.reset_index(inplace=True)

matching_version = 'title'
amazon_column_name = '제품명'
checco_column_name = 'productTitle'

# 컬럼명 맞춰줘야 해요
df1 = amazon_df.drop_duplicates(subset=[amazon_column_name])
df2 = checco_df.drop_duplicates(subset=[checco_column_name])
# Assuming you have loaded the datasets into two DataFrames: df1 and df2

for index, row1 in tqdm(df1.iterrows(), total= df1.shape[0]):
    max_score = 70
    # 컬럼명 맞춰줘야 해요
    product_name1 = row1[amazon_column_name]
    category_name1 = row1["카테고리명"]
    brand_name1 = row1["브랜드명"]
    if type(product_name1) == float:
        continue
    for index, row2 in df2.iterrows():
        # 컬럼명 맞춰줘야 해요
        product_name2 = row2[checco_column_name]
        category_name2 = row2["categoryName"]
        brand_name2 = row2["brandName"]
        if type(product_name2) == float:
            continue

            
            
        similarity_product_score = fuzz.ratio(product_name1, product_name2)
        similarity_category_score = fuzz.ratio(category_name1, category_name2)
        similarity_brand_score = fuzz.ratio(brand_name1, brand_name2)
        
        if similarity_category_score >= 90:
            df1.at[index, 'category_name'] = row2["categoryName"]
            if similarity_brand_score >= 90:
                df1.at[index, 'checco_brand_name'] = row2["brandName"]
        if similarity_brand_score >= 90:
            df1.at[index, 'checco_brand_name'] = row2["brandName"]
            if similarity_category_score >= 90:
                df1.at[index, 'category_name'] = row2["categoryName"]
        else:
            continue
        
        # Set a threshold value for similarity score above which two product names are considered a match
        if similarity_product_score >= 90:
            print("product_name1", product_name1)
            print("product_name2", product_name2)
            print("similarity_product_score", similarity_product_score)
            
            df1.at[index, 'Checco_product_id'] = row2["productId"]
            df1.at[index, 'Checco_product_title'] = product_name2
            df1.at[index, 'product_url'] = row2["productURL"]
            df1.at[index, 'image_url'] = row2["imageURL"]
            df1.at[index, 'similarity_product_score'] = similarity_product_score
            break
        elif similarity_product_score > max_score:  # Adjust the threshold as per your requirement
            max_score = similarity_product_score
            # Update the product name in df1 with the matched product name from df1
            df1.at[index, 'Checco_product_id'] = row2["productId"]
            df1.at[index, 'Checco_product_title'] = product_name2
            df1.at[index, 'product_url'] = row2["productURL"]
            df1.at[index, 'image_url'] = row2["imageURL"]
            df1.at[index, 'similarity_product_score'] = similarity_product_score

    # df1.to_csv(f'fuzzywuzzy_{matching_version}_drop.csv', index=False, encoding='utf-8-sig')
```
